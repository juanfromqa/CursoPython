```python
c = 0
while c <= 5:
    c+=1 
    print("c vale", c)
else:
    print("Se ha completado toda la iteración") # Esta linea se ejecuta solo cuando ya no se cumple la condición del while
    
```

```python
c = 0
while c <= 5:
    c+=1 
    print("c vale", c)
    if c == 3:
        print("SE INTERRUMPE LA EJECUCION EN", c)
        break
else:
    print("Se ha completado toda la iteración") # Esta linea ya no se ejecuta si se interrumpe el flujo
    
```

```python
c = 0
while c <= 5:
    c+=1 
    print("c vale", c)
    if c == 3:
        print("Continuamos con la siguiente iteración", c)
        continue
        print("Este print ya no se ejecuta -----")
else:
    print("Se ha completado toda la iteración") # Esta linea si se ejecuta con CONTINUE
    
```

```python
print("Bienvenido al menú interactivo")
while(True):
    print("""¿Qué quieres hacer? Escribe una opción
    1) Saludar
    2) Sumar dos números
    3) Salir""")
    opcion = input()
    if opcion == '1':
        print("Hola, espero que te lo estés pasando bien")
    elif opcion == '2':
        n1 = float(input("Introduce el primer número: "))
        n2 = float(input("Introduce el segundo número: "))
        print("El resultado de la suma es: ",n1+n2)
    elif opcion =='3':
        print("¡Hasta luego! Ha sido un placer ayudarte")
        break
    else:
        print("Comando desconocido, vuelve a intentarlo")
```
